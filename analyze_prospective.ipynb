```python
import pandas as pd
from os.path import join
from pytt.utils import read_pickle
instances = pd.read_csv(join('validation_annotation_results', 'annotations.data'), compression='gzip')
hierarchy = read_pickle(join('validation_annotation_results', 'hierarchy.pkl'))
```

```python
instances
```

```python
import json
model_counts = {}
for i,row in instances.iterrows(): # iterate over instances
    annotations = json.loads(row.annotations)
    for k1,v1 in annotations.items(): # iterate over annotators
        prev_tags = None
        for k2,v2 in v1.items(): # iterate over tabs
            if (k2 != 'future-reports') and (k2 != 'past-reports'):
                tags = set([])
                if k2 not in model_counts.keys():
                    model_counts[k2] = [0, 0]
                for k3,v3 in v2['tag_sentences'].items(): # iterate over tags/summaries
                    tags.add(k3)
                    for j in v3: # iterate over sentences
                        key = '_'.join([str(j),k3])
                        if key in v2['checked_senttags'].keys() and v2['checked_senttags'][key]:
                            model_counts[k2][0] += 1
                        model_counts[k2][1] += 1
                if prev_tags is not None and prev_tags != tags:
                    print(i, prev_tags, tags)
                prev_tags = tags
```
